```
import pandas as pd
```

```
df = pd.read_csv('/content/drive/MyDrive/Project/Hel_ri_celus (AI voice_bot for delivery_riders)/delivery_rider_driving_recognition/data/sample_gyr.csv')
df
```

```
df['time']
```

```
 df[['X_change','Y_change','Z_change']] = float(0)
```

```
for i in range(1,len(df),1):
    df['X_change'][i] = df['X_value'][i] - df['X_value'][i-1]
    df['Y_change'][i] = df['Y_value'][i] - df['Y_value'][i-1]
    df['Z_change'][i] = df['Z_value'][i] - df['Z_value'][i-1]
```

```
df
```

```
import seaborn as sns
df2 = df
df2.set_index(['time'],inplace=True)
sns.lineplot(data = df2[['X_value','Y_value','Z_value']])
```

```
# 변화량
sns.lineplot(data = df2[['X_change','Y_change','Z_change']])
```

```
# 표준 정규화

df3 = df

from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import numpy as np


Stan_scaler = StandardScaler()

df3 = Stan_scaler.fit_transform(df3)

sns.lineplot(data = df3[['X_change','Y_change','Z_change']])
```

```
# Min_Max 정규화

df4 = df2

Min_Max_scaler = MinMaxScaler()

df4[:] = Min_Max_scaler.fit_transform(df3[:])

sns.lineplot(data = df4[['X_change','Y_change','Z_change']])
```
